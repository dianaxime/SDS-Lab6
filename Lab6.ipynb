```python
def filter_set(dictionary, search_string):
    def iterator_func(x):
        if search_string in list(x.keys()):
            return True
        return False
    return filter(iterator_func, dictionary)
```

```python
filtered_records = filter_set(jsons_array, "dns")
filtered_records = list(filtered_records)
```

```python
from urllib.parse import urlparse, urlencode
```

```python
def getDomain(url):
    return urlparse(url).netloc
    
def getProtocol(url):
    return 1 if urlparse(url).scheme == 'https' else 0
    
def getSpecialCharacters(url):
    count_characters = 0
    count_characters += url.count('.')
    count_characters += url.count('-')
    count_characters += url.count('@')
    count_characters += url.count('?')
    count_characters += url.count('&')
    count_characters += url.count('|')
    count_characters += url.count('=')
    count_characters += url.count('_')
    count_characters += url.count('~')
    count_characters += url.count('%')
    count_characters += url.count('/')
    count_characters += url.count('*')
    count_characters += url.count(':')
    count_characters += url.count(',')
    count_characters += url.count(';')
    count_characters += url.count('$')
    count_characters += url.count('%20')
    count_characters += url.count(' ')
    return count_characters
```

```python
nuevo_df = pd.DataFrame()
```

```python
nuevo_df['longitud_url'] = unique_domains['dns.rrname'].str.len()
nuevo_df['ratio_digits_url'] = unique_domains['dns.rrname'].str.count('[0-9]') / unique_domains['dns.rrname'].str.len()
nuevo_df['ratio_digits_domain'] = unique_domains['dns.rrname'].apply(getDomain).str.count('[0-9]') / unique_domains['dns.rrname'].apply(getDomain).str.len()
```

```python
import whois
```

```python
def creation_date(domain):
    w = whois.whois(domain)
    
    return w.creation_date
```

```python
print(creation_date('webscraping.com'))
```
