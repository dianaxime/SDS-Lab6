```python
import json
import pandas as pd
```

```python
# Opening JSON file
f = open('large_eve.json')

info = f.readlines()

print(len(info))
# Closing file
f.close()
```

```python
print(info[0:10])
```

```python
jsons_array = [json.loads(line) for line in info if line]
```

```python
print(jsons_array[0:10])
```

```python
def filter_set(dictionary, search_string):
    def iterator_func(x):
        if search_string in list(x.keys()):
            return True
        return False
    return filter(iterator_func, dictionary)
```

```python
filtered_records = filter_set(jsons_array, "dns")
filtered_records = list(filtered_records)
```

```python
print(len(filtered_records))
```

```python
print(filtered_records[0:10])
```

```python
df = pd.json_normalize(filtered_records)
```
